```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
r_initial = np.array([-0.5, 0, 0])
v_initial = np.array([0.283/np.sqrt(2), 0.283/np.sqrt(2), 0])
w = np.array([0, 0, 1])
a_initial = -np.cross(w, np.cross(w, r_initial)) - 2*np.cross(w, v_initial)
r = np.linalg.norm(r_initial)
R = 1
```

```python
t = np.linspace(0, 20, 100000)
r_eff = r_initial
v_eff = v_initial
a_eff = a_initial
x = [r_eff[0]]
y = [r_eff[1]]
l = len(t)
i = -1
```

```python
while r < 1 and i < l:
    i += 1
    r_eff = r_eff + (t[i+1] - t[i])*v_eff
    v_eff = v_eff + (t[i+1] - t[i])*a_eff
    x.append(r_eff[0])
    y.append(r_eff[1])
    a_eff = -np.cross(w, np.cross(w, r_eff)) - 2*np.cross(w, v_eff)
    r = np.linalg.norm(r_eff)
```

```python
shp=patches.Circle((0,0), radius=R, color='r', fill=False)
plt.plot(x, y, 'b')
plt.gca().add_patch(shp)
plt.text(0, 0, '+')
plt.text(0, 0.25, 'v_0 = {0}'.format(np.linalg.norm(v_initial)) )
plt.text(0, -0.25, 'T = {0}'.format(t[i]))
plt.show()
```
