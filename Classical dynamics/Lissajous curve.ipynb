```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-2000*np.pi/2469, 2000*np.pi/2469, 10000)
```

```python
A = 1
B = 1
w_x = 4.9876
w_y = 1.2345
alpha = 0
beta = 0
```

```python
x = A*np.sin(w_x*t + alpha)
y = B*np.cos(w_y*t + beta)
```

```python
plt.plot(x, y)
plt.show()
```
